#### Muestra aleatorias

```python
def CrearDataSet(Number=1):
    Output = []
    for i in range(Number):
        #Crear rango de fechas semanal( de L a L) 
        rng = pd.date_range(start='1/1/2016', end='12/31/2030', freq='W-MON')
        #Crear valores aletarios
        data = np.random.randint(low=25, high= 100, size=len(rng))
        #Estatus posibles
        status = [1,2,3]
        #lista de estatius aleatorios
        random_status = [status[np.random.randint(low=0,high=len(status))]for i in range(len(rng))]
        #Estatus posibles
        states = ['Libertador', 'El hatillo','El Hatillo','Chacao','Baruta','Sucre']
        #Crear lista aleatoria de estados
        random_states = [states[np.random.randint(low=0,high=len(states))] for i in range(len(rng))]
        #extend permite agregar valores a una lista, similarmente a append
        Output.extend(zip(random_states,random_status,data,rng))
    return Output
        
```

```python
dataset = CrearDataSet(4)
df = pd.DataFrame(data=dataset,columns=['Local','Estatus_local','Cantidad_clientes','Fecha_Status'])
df
```

#### Operaciones sobre datos agrupados

```python
ratings = pd.read_csv(r'C:\Users\xiada\Desktop\ratings.csv')
peliculas = pd.read_csv(r'C:\Users\xiada\Desktop\peliculas.csv')
usuarios = pd.read_csv(r'C:\Users\xiada\Desktop\usuarios.csv')
```

```python
ratings.columns
```

```python
peliculas.columns
```

```python
usuarios.columns
```

```python
#Es posible que un usuario este asociado con un 0 o varios rating y
#películas, de la misma forma una película puede tener muchos ratings o 0,
#por varios usuarios.
peli_ratings = pd.merge(peliculas,ratings)
clasif = pd.merge(peli_ratings,usuarios)
```

```python
mas_ratings = clasif.groupby('titulo').size().sort_values(ascending=False)[:25]
mas_ratings
```

#### .agg()

```python
peli_stats = clasif.groupby('titulo').agg({'rating':[np.size,np.mean]}) 
peli_stats.head()
```

```python
peli_stats.sort_values([('rating','mean')], ascending=False).head()
```

```python
minimo_100 = peli_stats['rating']['size'] >= 100
```

```python
minimo_100 #Variables pandas que solo regrersa valores boleanos
```

```python
peli_stats[minimo_100].sort_values([('rating','mean')], ascending=False)[:15]
```
